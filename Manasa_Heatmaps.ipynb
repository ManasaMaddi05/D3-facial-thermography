### This heatmap visualizes how combinations of forehead center temperature (T_FHCC1) and right canthus temperature (T_RC1) relate to a person’s actual oral temperature, revealing regions of strong predictive alignment.

```python
# Extract and clean relevant data
df = FLIR_group1[['T_FHCC1', 'T_RC1', 'aveOralM']].dropna()

# Bin the temperature values
x_bins = pd.cut(df['T_FHCC1'], bins=20)
y_bins = pd.cut(df['T_RC1'], bins=20)

# Create a pivot table to calculate average oral temperature for each bin combo
heatmap_data = df.pivot_table(
    index=y_bins, columns=x_bins, values='aveOralM', aggfunc='mean'
)

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, cmap='coolwarm', cbar_kws={'label': 'Avg Oral Temp (°C)'})
plt.title("Oral Temperature by Forehead vs Right Canthus Temps")
plt.xlabel("T_FHCC1 (Forehead Center, Binned)")
plt.ylabel("T_RC1 (Right Canthus, Binned)")
plt.tight_layout()
plt.show()
```

### Findings
- Higher combinations of both T_FHCC1 and T_RC1 are associated with elevated oral temperatures, as seen in the warmer (red) bottom-right region
- Lower values for both regions correspond to lower oral temperatures
- The most accurate prediction zones for oral temperature appear when both facial regions are elevated — suggesting that paired facial readings may enhance fever screening.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assume your DataFrames are already loaded as follows:
# FLIR_group1 = pd.read_csv("cleaned_FLIR_group1.csv", header=2)
# FLIR_group2 = pd.read_csv("cleaned_FLIR_group2.csv", header=2)

# Drop any unwanted unnamed columns
FLIR_group1 = FLIR_group1.loc[:, ~FLIR_group1.columns.str.contains("^Unnamed")]
FLIR_group2 = FLIR_group2.loc[:, ~FLIR_group2.columns.str.contains("^Unnamed")]

# Define the variables of interest for both groups
vars_of_interest = ['T_FHCC1', 'T_RC1', 'aveOralM']

# Filter and drop rows with missing values
df1 = FLIR_group1[vars_of_interest].dropna()
df2 = FLIR_group2[vars_of_interest].dropna()

# Bin the facial temperature data for group 1
x_bins1 = pd.cut(df1['T_FHCC1'], bins=20)
y_bins1 = pd.cut(df1['T_RC1'], bins=20)

# Create a pivot table: average oral temp for each bin combination for group 1
heatmap_data1 = df1.pivot_table(index=y_bins1, columns=x_bins1, values='aveOralM', aggfunc='mean')

# Bin the facial temperature data for group 2
x_bins2 = pd.cut(df2['T_FHCC1'], bins=20)
y_bins2 = pd.cut(df2['T_RC1'], bins=20)

# Create a pivot table: average oral temp for each bin combination for group 2
heatmap_data2 = df2.pivot_table(index=y_bins2, columns=x_bins2, values='aveOralM', aggfunc='mean')

# Plot both heatmaps side-by-side for comparison
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

sns.heatmap(heatmap_data1, cmap='coolwarm', cbar_kws={'label': 'Avg Oral Temp (°C)'}, ax=axes[0])
axes[0].set_title("Group 1: FLIR (Ambient 20–24°C)")
axes[0].set_xlabel("T_FHCC1 (Forehead Center, Binned)")
axes[0].set_ylabel("T_RC1 (Right Canthus, Binned)")

sns.heatmap(heatmap_data2, cmap='coolwarm', cbar_kws={'label': 'Avg Oral Temp (°C)'}, ax=axes[1])
axes[1].set_title("Group 2: FLIR (Ambient 24–29°C)")
axes[1].set_xlabel("T_FHCC1 (Forehead Center, Binned)")
axes[1].set_ylabel("T_RC1 (Right Canthus, Binned)")

plt.tight_layout()
plt.show()
```

This side-by-side heatmap comparison explores whether the relationship between facial temperatures (forehead center and right canthus) and oral temperature is consistent across different ambient temperature groups (20–24°C vs. 24–29°C). It reveals that ambient conditions do subtly shift the distribution and magnitude of temperature readings

Similar overall trend across both groups: In both heatmaps, higher values for forehead and canthus temperatures tend to correspond to higher oral temperatures

Differences:
- In Group 2 (24–29°C), there’s a shift in the binned temperature ranges — they are slightly higher overall, likely reflecting the influence of a warmer ambient environment on surface temperature readings
- This means that to detect the same core temperature, facial readings are inflated in warmer environments
